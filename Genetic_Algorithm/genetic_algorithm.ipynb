```python
import math
import random
import threading
from collections import defaultdict, deque
```

```python
V = int(input("No. of vertices: "))
```

```python
def generate_random_edges(V):
    edges = [ (u, v, random.randint(1, 100)) for u in range(1, V)
             for v in range(u+1, V+1)]
    return edges
```

```python
edges = generate_random_edges(V)
```

```python
edges
```

```python
E = len(edges)
```

```python
start, end = map(int, input("start, end:").split(" "))
```

```python
C = int(input("No. of candidates: "))
```

```python
cands = [[0]*E for i in range(C)]
```

```python
def assign_vals(cand, V):
    idx = random.sample(range(len(cand)), V-1)
    for i in idx:
        cand[i] = 1
```

```python
threads = []
for c in cands:
    th = threading.Thread(target=assign_vals, args=(c, V))
    threads.append(th)
    th.start()
for th in threads:
    th.join()
print(*cands, sep='\n')
```

```python
def get_fitness(path, start, end, vis):
    if start == end:
        return 0
    vis.add(start)
    min_fit = float('inf')
    for v, w in path[start]:
        if v in vis:
            continue
        min_fit = min(min_fit, get_fitness(path, v, end, vis) + w)
    return min_fit
```

```python
def check_path(cand, edges, start, end, fits):
    path = defaultdict(list)
    for c in range(len(cand)):
        if cand[c]:
            u, v, w = edges[c]
            path[u].append((v, w))
            path[v].append((u, w))
    if end not in path:
        fits.append(float('inf'))
        return
    fits.append(get_fitness(path, start, end, set()))
            
```

```python
l = []
check_path(cands[1], edges, start, end, l)
l
```

```python
fitness = []
for c in cands:
    check_path(c, edges, start, end, fitness)
    print(*c, "fitness: %.2f"%fitness[-1])
```

```python
def do_crossover(cands, fits):
    sorted_cands = [i[0] for i in sorted(zip(cands, fits), key = lambda i:i[1])]
    new_cands = []
    S = len(sorted_cands)
    for i in range(S//2):
        pt = random.randint(1, E-2)
        new = sorted_cands[i][:pt] + sorted_cands[S-i-1][pt:]
        new_cands.append(new)
    return cands
```

```python
min_fitness = float('inf')
count = 0
for i in range(1000):
    fits = list()
    threads = list()
    for c in cands:
        th = threading.Thread(target=check_path, args=(c, edges, start, end, fits))
        threads.append(th)
        th.start()
    for th in threads:
        th.join()
    min_fit = min(fits)
    if math.isclose(min_fit, min_fitness, rel_tol=0.1):
        count += 1
    else:
        count = 0
    if i%10 == 0:
        print("fitness: %d"%min_fit)
    min_fitness = min(min_fitness, min_fit)
    if count == 10:     # if repeated value for 10 iters than break
        break
    cands = do_crossover(cands, fits)
```
